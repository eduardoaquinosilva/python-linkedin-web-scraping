```python
jobs = list()

# run through all jobs, collect their data and append it to a dicts list
for vacancy in vacancies_list:
    job_title = vacancy.select(".base-search-card__title")[0].text.strip()  # job title
    print(f"Job: {job_title}")
    
    company = vacancy.select(".base-search-card__subtitle")[0].text.strip()  # job company
    print(f"Company: {company}")
    
    location = vacancy.select(".job-search-card__location")[0].text.strip()  # job's location
    print(f"Location: {location}")
    
    posting_date = vacancy.select("time")[0].get("datetime")  # when posted
    print(f"Date of Post: {posting_date}")
    
    if len(vacancy.select(".hidden-nested-link")) == 0:
        job_link = vacancy.select(".base-card")[0].get("href")
    else:
        job_link = vacancy.select(".hidden-nested-link")[0].get("href")
    print(f"Link: {job_link}")

    print("########\n\n\n")
    
    info = {
        "Job": job_title,
        "Company": company,
        "Location": location,
        "Date of Post": posting_date,
        "Link": job_link
    }
    
    jobs.append(info)
```

```python
jobs
```

```python
json_jobs = json.dumps(jobs)
json_jobs
```

```python
# create json file and fill it
with open("jobs.json", "w") as json_file:
    json_file.write(json_jobs)
```

```python
# create csv file and fill it
with open('jobs.csv', mode='w') as csv_file:
    header = ['Job', 'Company', 'Location', "Date of Post", "Link"]
    generate_csv = csv.DictWriter(csv_file, fieldnames=header)
    generate_csv.writeheader()
    generate_csv.writerows(jobs)
```
